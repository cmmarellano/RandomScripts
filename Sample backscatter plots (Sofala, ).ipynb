```python
import pandas as pd
```

```python
df_sofala00 = pd.read_csv('sofala00.csv', sep=';')
print(df_sofala00)
```

```python
type(df_sofala00)
```

```python
df_sofala00.apply(lambda row: row.VH / row.VV, axis=1)
```

```python
df_sofala00['VH/VV'] = df_sofala00.apply(lambda row: row.VH / row.VV, axis=1)
```

```python
print(df_sofala00.describe())
```

```python
df_sofala00.plot('date', 'VH')
df_sofala00.plot('date', 'VV')
df_sofala00.plot('date', 'VH/VV')
```

```python
import matplotlib.pyplot as plt

# gca stands for 'get current axis'
ax = plt.gca()

df_sofala00.plot(x='date', y='VH', ax=ax)
df_sofala00.plot(x='date', y='VV', ax=ax)
df_sofala00.plot(x='date', y='VH/VV', ax=ax)



# Increase size of plot in jupyter
plt.rcParams["figure.figsize"] = (18,6)


plt.title('Sofala sample plot')
plt.ylabel('sig0 (db)')
plt.tight_layout()
plt.show()
```

```python
col1, col2 = "VH", "VV"
corrVHVV = df_sofala00['VH'].corr(df_sofala00['VV'])
print("Correlation between {} and {} is {}".format(col1,col2,round(corrVHVV, 2)))

col3, col4 = "VH", "VH/VV"
corrVHCR = df_sofala00['VH'].corr(df_sofala00['VH/VV'])
print("Correlation between {} and {} is {}".format(col3,col4,round(corrVHCR, 2)))

col5, col6 = "VV", "VH/VV"
corrVVCR = df_sofala00['VV'].corr(df_sofala00['VH/VV'])
print("Correlation between {} and {} is {}".format(col5,col6,round(corrVVCR, 2)))
```
